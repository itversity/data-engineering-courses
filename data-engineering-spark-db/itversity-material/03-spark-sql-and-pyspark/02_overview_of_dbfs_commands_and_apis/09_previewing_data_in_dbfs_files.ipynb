## Previewing data in DBFS Files

Let us see how we can preview the data in DBFS using Notebook Environment.

```python
display(spark.read.json('/public/retail_db_json/orders'))
```

```python
display(spark.sql('SELECT * FROM JSON.`/public/retail_db_json/orders`'))
```

```python
%sql

SELECT * FROM JSON.`/public/retail_db_json/orders`
```

```python
%fs ls /databricks-datasets/airlines
```

```python
%sql

SELECT * FROM TEXT.`/databricks-datasets/airlines/part-00000`
```

```python
%sql

CREATE TEMPORARY VIEW airlines
USING CSV
OPTIONS (
    path='/databricks-datasets/airlines/part-00000',
    header=True,
    inferSchema=True
)
```

```python
%sql

SELECT * FROM airlines
```
