## Copying files from DBFS to DBFS using fs magic

We can copy files DBFS HDFS to DBFS either by using `databricks fs cp` on our local environment or `%fs cp` using Databricks Notebooks. We will primarily focus on `%fs cp`

* `%fs cp` can be used with `-r` to copy a folder recursively from one DBFS location to another DBFS location.

```python
%fs cp /public/retail_db /public/retail_db_copy -r
```

```python
%fs ls /public/retail_db_copy
```

```python
dbutils.fs.ls('/public/retail_db_copy')
```

```python
dbutils.fs.help('ls')
```

```python
list(map(lambda rec: (rec.name, rec.isDir()), dbutils.fs.ls('dbfs:/public/retail_db_copy')))
```

```python
%fs rm /public/retail_db_copy -r
```
