## Getting File Metadata

Let us see how to get metadata for the files stored in DBFS using `dbutils.fs.ls` command.

```python
dbutils.fs.ls('/databricks-datasets/airlines')
```

```python
# Get number of files which start with part.

list(filter(lambda rec: rec.name.startswith('part'), dbutils.fs.ls('/databricks-datasets/airlines')))
```

```python
len(list(filter(lambda rec: rec.name.startswith('part'), dbutils.fs.ls('/databricks-datasets/airlines'))))
```

```python
# Get overall size in Gigabytes

list(map(lambda rec: rec.size, filter(lambda rec: rec.name.startswith('part'), dbutils.fs.ls('/databricks-datasets/airlines'))))
```

```python
round(sum(map(lambda rec: rec.size, filter(lambda rec: rec.name.startswith('part'), dbutils.fs.ls('/databricks-datasets/airlines')))) / (1024 * 1024 * 1024), 2)
```
