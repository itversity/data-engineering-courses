## Managing DBFS Directories using fs magic or dbutils

Now let us have a look at how to manage DBFS directories using databricks `fs` magic or `dbutils`.

* `%fs mkdirs` or `dbutils.fs.mkdirs`â€“ to create directories.
* Let's go ahead and create directory by name **dummy** under **dbfs:/public**.

```python
%fs mkdirs dbfs:/public/dummy
```

```python
%fs ls dbfs:/public
```

* Let us delete the empty directory using below command. We can use `%fs rm` command with `-r` option to delete non-empty directory or folder recursively.

```python
%fs rm dbfs:/public/dummy -r
```

```python
dbutils.fs.help('mkdirs')
```

```python
dbutils.fs.mkdirs('dbfs:/public/dummy')
```

```python
%fs ls dbfs:/public
```

```python
dbutils.fs.help('rm')
```

```python
dbutils.fs.rm('dbfs:/public/dummy', recurse=True)
```
