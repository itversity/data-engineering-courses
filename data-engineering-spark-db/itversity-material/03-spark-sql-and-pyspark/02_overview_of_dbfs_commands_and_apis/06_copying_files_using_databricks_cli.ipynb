## Copying files using databricks CLI

We can copy files from local file system from our PC or Mac to DBFS by using `databricks fs cp` command. We can also use the same command to copy from DBFS to local file system on our PC or Mac.

* `databricks fs cp` have very limited options such as `-r` or `--recursive`.
* It is primarily used to copy smaller data sets for proof of concepts or prototypes.
* The `databricks fs cp` command might take long time even to copy moderate size files or folders.
* Typically we use tools or custom solutions to ingest data into DBFS in actual projects.
* Already, we have used to copy **data/retail_db** folder recursively into **dbfs:/public/retail_db**.
* Take it as an exercise and try copying data from **data/retail_db_json** folder recursively into **dbfs:/public/retail_db_json**.
* You can run `databricks fs ls dbfs:/public/retail_db_json` to validate if all the files and folders are successfully copied or not.
