## Getting help or usage for DBFS Commands

Let us explore details about how to get the help or usage for DBFS Commands. We can run Databricks File System commands using `%fs` as part of Databricks Notebooks.
* Run below cell to see the usage of `%fs` magic.

```python
%fs
```

We can use `%fs ls` to list the files. To get help we can use `dbutils.fs.help(ls)` or `%fs help ls`.

```python
%fs help ls
```

```python
dbutils.fs.help('ls')
```

```python
%fs ls
```

```python
%fs rm dbfs:/dummy -r
```

```python
dbutils.fs.ls('dbfs:/')
```

```python
%fs ls dbfs:/public/retail_db
```

```python
dbutils.fs.ls('dbfs:/public/retail_db')
```
