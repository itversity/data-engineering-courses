## Using Databricks CLI to Manage DBFS

Let us revise details about using Databricks CLI. We have already gone through the details earlier once. You need to use this Notebook in local project setup in your Mac or Windows.
* You need to ensure Databricks CLI is installed and configured.
* You can get the usage of Databricks CLI Command using `databricks -h`.
* Following are most common scenarios where **databricks cli** comes handy.
  * Managing Data Files using `databricks fs`
  * Managing Notebooks, Dependencies, etc using `databricks workspace`
  * Managing Clusters using `databricks clusters`
  * Managing Databricks Jobs using `databricks jobs`
* Here are the examples related to Managing Files in fs.

```python
!databricks fs ls # to list files
```

```python
!databricks fs mkdirs dbfs:/dummy # create folder by name dummy
```

```python
!databricks fs cp data/retail_db_json dbfs:/public/retail_db_json -r # to copy retail_db_json folder into DBFS recursively
```

```python
!databricks fs rm dbfs:/public/retail_db_json -r
```
