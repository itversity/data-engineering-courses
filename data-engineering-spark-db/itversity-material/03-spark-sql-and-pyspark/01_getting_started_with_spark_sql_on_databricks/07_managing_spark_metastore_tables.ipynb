## Managing Spark Metastore Tables

Let us create our first Spark Metastore table. We will also have a look into how to list the tables.

* We will get into details related to DDL Commands at a later point in time. 
* For now we will just create our first table. We will get into the details about creating tables as part of subsequent sections.

```scala
%sql
SELECT current_database()
```

```scala
%sql
SHOW databases
```

```scala
%sql
DROP DATABASE IF EXISTS itversity_retail CASCADE
```

```scala
%sql
SHOW databases
```

```scala
%sql
CREATE DATABASE IF NOT EXISTS itversity_retail
```

```scala
%sql
USE itversity_retail
```

```scala
%sql
SELECT current_database()
```

```scala
%sql
CREATE TABLE orders (
    order_id INT,
    order_date STRING,
    order_customer_id INT,
    order_status STRING
) ROW FORMAT DELIMITED FIELDS TERMINATED BY ','
```

* We can list the tables using `SHOW tables;`

```scala
%sql
SHOW tables
```

```scala
%sql
SELECT * FROM orders
```

```scala
%sql
SELECT count(*) FROM orders;
```

```scala
%sql
DESCRIBE FORMATTED orders
```

```scala
%fs ls /public/retail_db/orders
```

```scala
%fs cp /public/retail_db/orders/part-00000 /user/hive/warehouse/itversity_retail.db/orders/part-00000
```

```scala
%sql
SELECT * FROM orders
```

```scala
%sql
SELECT count(*) FROM orders
```

* We can also drop the table using `DROP TABLE` command. We will get into more details at a later point in time.
* We can also truncate the managed tables using `TRUNCATE TABLE` command.

```scala
%sql
DROP TABLE itversity_retail.orders
```

```scala
%fs ls /user/hive/warehouse/itversity_retail.db
```

```scala
%sql
DROP DATABASE IF EXISTS itversity_retail CASCADE
```
